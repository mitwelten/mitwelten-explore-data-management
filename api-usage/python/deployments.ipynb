```python
import requests
import datetime
import pandas as pd
import numpy as np
import plotly.graph_objects as go
```

```python
BASE_URL = "https://data.mitwelten.org/api/v3/"
deployments = requests.get(BASE_URL+"deployments").json()
```

```python
lats = [d.get("location").get("lat") for d in deployments]
lons = [d.get("location").get("lon") for d in deployments]
nodes = [d.get("node") for d in deployments]
```

```python
fig = go.Figure()
fig.add_trace(go.Scattermapbox(lat=lats,lon=lons, mode="markers"))
fig.update_layout(
    mapbox = {
        'style': "stamen-terrain",
        'center': {'lon': np.mean(lons), 'lat': np.mean(lats)},
        'zoom': 11},
        margin=dict(l=0,r=0,t=0,b=0),
    showlegend = False)
fig.show()
```
